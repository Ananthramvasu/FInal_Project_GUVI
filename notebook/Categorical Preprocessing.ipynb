```python
# Convert column names to lowercase
df_ob.columns = df_ob.columns.str.lower()

# Capitalize the first letter of each word and keep underscores
df_ob.columns = df_ob.columns.str.title()
df_ob.columns
```

```python
'''Highly Relevant Columns
- Policy_Id – Unique identifier for each policy (for tracking claims).
- Region_Code – Geographic location can impact risk (e.g., accident-prone areas).
- Segment – Helps categorize policyholders based on risk factors.
- Model – The type of vehicle can influence insurance premiums and risk.
- Fuel_Type – Certain fuel types may have different risk levels (e.g., electric vs. diesel).
- Max_Torque / Max_Power – Indicators of vehicle performance, which may influence risk.
- Engine_Type – Engine reliability and efficiency can impact likelihood of claims.
- Claim_History_Label – Past claims history is crucial for risk assessment.
- Risk_Score – A direct measure of insurance risk.


Moderately Relevant Columns (Safety Features & Vehicle Attributes)
- Is_Esc (Electronic Stability Control) – Enhances vehicle safety, reducing accident risk.
- Is_Adjustable_Steering – Could impact ease of vehicle control.
- Is_Tpms (Tire Pressure Monitoring System) – Reduces accident probability.
- Is_Parking_Sensors / Is_Parking_Camera – Helps assess risk related to parking incidents.
- Rear_Brakes_Type / Transmission_Type / Steering_Type – Contribute to handling and accident likelihood.
- Is_Front_Fog_Lights / Is_Rear_Window_Wiper / Is_Rear_Window_Defogger / Is_Rear_Window_Washer – Affect visibility in adverse weather conditions.
- Is_Brake_Assist / Is_Power_Door_Locks / Is_Central_Locking / Is_Power_Steering – Impacts safety and security.
- Is_Driver_Seat_Height_Adjustable / Is_Day_Night_Rear_View_Mirror – Comfort and visibility factors.
- Is_Ecw (Early Crash Warning) / Is_Speed_Alert – Reduces risk of crashes.


Less Relevant Columns
- Gender – May be used in some underwriting models, though modern practices focus more on behavior and history.
- Policy_Type – Determines coverage details but does not directly affect risk.
- Other safety features – These may have an indirect effect, but their weight in risk modeling depends on historical data.'''
```

```python
#After Breaking down into the these categories let us come to conclusion that we can omit moderately Relevant Columns because it is of vehicle manufacturing.
#Since we focus on insuring part we now consider on the other two aspects.
```

```python
highly_relevant_columns = [
    'Policy_Id', 'Region_Code', 'Segment', 'Model', 'Fuel_Type','Max_Torque', 'Max_Power', 'Engine_Type', 'Claim_History_Label', 'Risk_Score']

less_relevant_columns = ['Gender', 'Policy_Type','Claim_History_Label']

# Combine all relevant columns
relevant_columns = highly_relevant_columns + less_relevant_columns

# Create a new DataFrame with only those columns
df_cat_cols = df_ob[relevant_columns].copy()
```

```python
#Now we are done with separating columns let us proceed with encoding.
```

```python
from sklearn.preprocessing import LabelEncoder

# List of categorical columns to encode
label_cols = [
    'Region_Code', 'Segment', 'Model', 'Fuel_Type',
    'Engine_Type', 'Claim_History_Label', 'Risk_Score',
    'Gender', 'Policy_Type'
]

# Encode each column one-by-one safely
for col in label_cols:
    if df_cat_cols[col].ndim == 1:  # ensure it's 1D
        le = LabelEncoder()
        df_cat_cols[col] = le.fit_transform(df_cat_cols[col].astype(str))
    else:
        print(f"Skipping column {col} due to unexpected shape: {df_cat_cols[col].shape}")
```

```python
# Remove duplicate columns (same name)
df_cat_cols = df_cat_cols.loc[:, ~df_cat_cols.columns.duplicated()]
```

```python
from sklearn.preprocessing import LabelEncoder

label_cols = [
    'Region_Code', 'Segment', 'Model', 'Fuel_Type',
    'Engine_Type', 'Claim_History_Label', 'Risk_Score',
    'Gender', 'Policy_Type'
]

for col in label_cols:
    le = LabelEncoder()
    df_cat_cols[col] = le.fit_transform(df_cat_cols[col].astype(str))
```
