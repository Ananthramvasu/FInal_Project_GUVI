```python
# Example using TextBlob
from textblob import TextBlob

def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

df['Sentiment_label'] = df['Review Text'].apply(get_sentiment)
```

```python
# Define a simple mapping
sentiment_to_rating = {
    'Positive': 5,
    'Neutral': 3,
    'Negative': 1
}

df['Review_rating'] = df['Sentiment_label'].map(sentiment_to_rating)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
df.columns
```

```python
X=df['Review_text']
y=df['Sentiment_label']
```

```python
cv = CountVectorizer()
X_cv = cv.fit_transform(X) # Fit the Data
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_cv, y)
```

```python
clf = MultinomialNB()
clf.fit(X_train,y_train)
```
