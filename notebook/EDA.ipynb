```python
import pandas as pd
import pickle
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_pickle('dataset.pkl')
```

```python
df1
```

```python
df2 = pd.read_pickle('Categorical_Dataset.pkl')
```

```python
df2.head()
```

```python
df = pd.concat([df1,df2],axis = 1)
```

```python
df.head()
```

```python
df.columns
```

```python
#univariate analysis
```

```python
df['Claim_Amount'].hist(bins=50)
sns.kdeplot(df['Annual_Income'])
```

```python
#Distribution of Insurance Premium using histplot
plt.figure(figsize=(6,4))
sns.histplot(df['Insurance_Premium'], bins=10, kde=True, color='brown')
plt.title("Distribution of Insurance Premium")
plt.xlabel("'Insurance Premium'")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='Risk_Score', data=df)
sns.countplot(x='Policy_Type', hue='Fraudulent_Claim', data=df)
```

```python
#BI variate analysis
```

```python
# lineplot Insurance Premium amount based on Engine Type
plt.figure(figsize=(6,4))
sns.lineplot(x=df['Engine_Type'], y=df['Insurance_Premium'], palette=["green", "red"])
plt.title("Insurance Premium amount based on Engine Type")
plt.xlabel("Engine Type")
plt.ylabel("Insurance Premiun")
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data=df,x='Gender',y='Region_Density')
plt.title("Gender based on Regions")
plt.xlabel("Gender")
plt.ylabel("Region Density")           
plt.show()
```

```python
sns.lmplot(data=df, x='Claim_Amount', y='Vehicle_Age', height=4, aspect=1.5)
plt.title("Claim Amount based on Vehicle Age")
plt.xlabel("Claim Amount")
plt.ylabel("Vehicle Age")
plt.show()
```

```python
pivot = df.pivot_table(values='Claim_Amount', 
                       index=pd.cut(df['Vehicle_Age'], 5),
                       columns=pd.cut(df['Claim_Amount'], 5), 
                       aggfunc='count')

sns.heatmap(pivot, cmap='YlGnBu')
plt.title("Heatmap of Vehicle Age vs Claim Amount Frequency")
plt.xlabel("Claim Amount Range")
plt.ylabel("Vehicle Age Range")
plt.show()
```

```python
sns.boxplot(x='Fraudulent_Claim', y='Claim_Amount', data=df)
sns.violinplot(x='Fraudulent_Claim', y='Annual_Income', data=df)
```

```python
pd.crosstab(df['Risk_Score'], df['Fraudulent_Claim'], normalize='index').plot(kind='bar', stacked=True)
```

```python
#multivariet analysis
```

```python
sns.pairplot(df[['Claim_Amount', 'Insurance_Premium', 'Vehicle_Age', 'Annual_Income', 'Fraudulent_Claim']], hue='Fraudulent_Claim')
```

```python
#Group-Based Insights
```

```python
df.groupby('Policy_Type')['Claim_Amount'].mean().plot(kind='bar')
```

```python
#Comparison Charts
```

```python
sns.scatterplot(x='Annual_Income', y='Claim_Amount', hue='Fraudulent_Claim', data=df, alpha=0.6)
plt.title("Income vs Claim Amount (colored by Fraud)")
```

```python
sns.barplot(x='Fuel_Type', y='Claim_Amount', data=df)
plt.title("Average Claim Amount by Fuel Type")
```
