```python
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
df=pd.read_pickle('final_dataframe.pkl')
df1 = pd.read_pickle('fraudent_claim.pkl')
```

```python
df1['Fraudulent_Claim'].value_counts()
```

```python
X = df[['Vehicle_Age', 'Customer_Age','Airbags', 'Displacement', 'Cylinder','Gross_Weight', 'Annual_Income', 'Insurance_Premium', 'Claim_History','Segment', 'Model', 'Fuel_Type','Claim_History_Label','Gender', 'Policy_Type', 'Max_Torque_Nm','Max_Power_bhp']]  # Drop both target columns
y = df1['Fraudulent_Claim']   # New target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
rfc.score(X_train,y_train)
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_test,y_test)
```

```python
dtc.score(X_train,y_train)
```

```python
knnc = KNeighborsClassifier()
knnc.fit(X_train,y_train)
```

```python
knnc.score(X_train,y_train)
```

```python
gbc = GradientBoostingClassifier()
gbc.fit(X_train,y_train)
```

```python
gbc.score(X_train,y_train)
```

```python
prediction = rfc.predict(X_test)
```

```python
for i in prediction:
    if i == 1:
        print("Fraudulent Claim (Fraud detected)")
    else:
        print("Genuine Claim (Not Fraud)")
```
