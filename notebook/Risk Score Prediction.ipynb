```python
X = df[['Vehicle_Age', 'Customer_Age','Airbags', 'Displacement', 'Cylinder','Gross_Weight', 'Annual_Income', 'Insurance_Premium', 'Claim_History','Segment', 'Model', 'Fuel_Type','Gender', 'Policy_Type', 'Max_Torque_Nm','Max_Power_bhp']]  # Drop both target columns
y = df['Risk_Score']   # New target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
rfc.score(X_train, y_train)
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_test,y_test)
```

```python
dtc.score(X_train,y_train)
```

```python
knnc = KNeighborsClassifier()
knnc.fit(X_train,y_train)
```

```python
knnc.score(X_train,y_train)
```

```python
gbc = GradientBoostingClassifier()
gbc.fit(X_train,y_train)
```

```python
gbc.score(X_train,y_train)
```

```python
prediction = rfc.predict(X_test)
```
